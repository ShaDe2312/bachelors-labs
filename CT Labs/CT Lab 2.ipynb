```python
import control
import numpy as np
import matplotlib.pyplot as plt

sys = control.TransferFunction(np.array([2,3,1]) , np.array([12,7,1]))
control.pzmap(sys, plot=True, grid=None, title='Pole Zero Map')
```

```python
#For RC Circuit with Capacitance of 1 mF and resistance of 100k Ohm
num=[1]
den=[100,1]
RC=control.tf(num,den)
print('H(s)=',RC)
control.pzmap(RC, plot=True, grid=None, title='Pole Zero Map')
```

```python
# Vout = 0.1 * np.exp(0.1 * t)

temp = control.impulse_response(RC, T=100, X0=0.0, input=None, output=None, T_num=1000, transpose=False, return_x=False, squeeze=None)

plt.plot(temp)
plt.show()
```

```python
#Impulse response of RC Circuit
a0=1
a1=10
b0=1
b1=0
num = np.array([b1, b0])
den = np.array([a1, a0])
H = control.tf(num, den)
print ('H ( s ) = ' , H )


t=np.arange(0,100.1,0.1)
T, yout1 = control.impulse_response(H, t, X0=0)

plt.plot(T,yout1)
plt.title("Impulse Response of RC Circuit")
plt.xlabel("Time")
plt.ylabel("Vout")
plt.show()
```

```python
#Step response of RC Circuit
H2= control.tf([1], [1,0.1])
T2, yout2 = control.step_response(H2, t, X0=0)

plt.plot(T2,yout2)
plt.title("Step Response of RC Circuit")
plt.xlabel("Time")
plt.ylabel("Vout")
plt.show()
```
