```python
import control
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python

a0 = 1
a1 = 0.001
b0 = 1
b1 = 0

# Transfer Function
num = np.array([b1 , b0])
den = np.array([a1 , a0])
H = control.tf(num,den)
print(H)

w = [1,100,200,300,400,500,600,700,800,900,1000,1200,1400,1600,1800,2000]

freq = []
for i in range(0, len(w)):
    a = w[i]/(2*math.pi)
    freq.append(a)
    
mag, phase, omega = control.freqresp(H, w)

plt.plot(freq, mag)
plt.title("Frequency Vs Magnitude")
plt.xlabel("w")	
plt.ylabel("h")
plt.show()

plt.plot(freq, phase)
plt.title("Frequency Vs Phase")
plt.xlabel("w")
plt.ylabel("phase")
plt.show()
```
