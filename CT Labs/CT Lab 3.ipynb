```python
import numpy as np
import control
from matplotlib import pyplot as plt
```

```python
# case 1 (zeta > 1) overdamped system
r = 4
l = 1
c = 1
a0= 1/(l*c)
a1= r/l
a2= 1
b0= 1/(l*c)
b1= 0

num = np.array([b1, b0])
den = np.array([a2, a1, a0])
H = control.tf(num, den)
print ('overdamped system')
print ('H ( s ) = ' , H )

control.pzmap(sys=H, plot=True, grid=False, title='Pole Zero Map -(zeta>1) - overdamped system')
plt.show()
t = np.arange(0,20,0.1)
x,y = control.step_response(H, t, X0=0)

plt.plot(x,y)
plt.grid(True)
plt.xlabel('Time')
plt.ylabel('Response')
plt.title('Step response of RLC circuit - (zeta>1) - overdamped system')
plt.show()
```

```python
# case 2 (0 < zeta < 1) underdamped system
r = 1
l = 1
c = 1
a0= 1/(l*c)
a1= r/l
a2= 1
b0= 1/(l*c)
b1= 0

num = np.array([b1, b0])
den = np.array([a2, a1, a0])
H = control.tf(num, den)

print ('underdamped system')
print ('H ( s ) = ' , H )
control.pzmap(sys=H, plot=True, grid=False, title='Pole Zero Map - (0 <zeta < 1) - underdamped system')
plt.show()

t = np.arange(0,20,0.1)
x,y = control.step_response(H, t, X0=0)

plt.plot(x,y)
plt.grid(True)
plt.xlabel('Time')
plt.ylabel('Response')
plt.title('Step response of RLC circuit - (0 < zeta < 1) - underdamped system')
plt.show()
```
