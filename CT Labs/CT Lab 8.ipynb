```python
import numpy as np 
import control 
from matplotlib import pyplot as plt 
from control import matlab

R = 240 
L = 40 
C1 = 0.001 
A = [[0,1/C1],[-1/L,-R/L]] 
B = [[0],[1/L]] 
C = [[1,0],[0,R]] 
D = [[0],[0]] 

S = control.ss(A,B,C,D) 
print(S) 
print("here")

tf = control.ss2tf(S) 
print(tf) 

T,y= control.step_response(tf[0,0]) 
plt.plot(T,y) 
plt.xlabel("Time")
plt.ylabel("Vout")
plt.show() 
```
