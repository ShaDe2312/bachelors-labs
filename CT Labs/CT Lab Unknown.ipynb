```python
import numpy as np
import control
import matplotlib.pyplot as plt
import math
```

```python
numerator=np.array([250])
denominator= np.array([1,12,20,0])

```

```python
def system(numerator,denominator):
    system1= control.tf(numerator, denominator)

    # %% Displaying the transfer function :
    print ('G(s)H(s) = ', system1)

    magnitude, phase, omega= control.bode(system1 , dB=True)
    plt.plot(omega)
    plt.plot(phase)
    plt.plot(magnitude)
    plt.show()

    gm, pm, wg, wp = control.margin(system1)
    print("Gain margin (db): ",20*log10(gm), 
          "\nPhase margin (degrees): ",pm*180/(math.pi), 
          "\nFrequency of gain margin: ",wg,
          "\nFrequency of phase margin: ", wp)
```
