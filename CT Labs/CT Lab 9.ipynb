```python
import control 
import matplotlib.pyplot as plt
```

```python

c = 0.001
L = 40
R = 240

A = [[0,1/c], [-1/L,-R/L]]
B = [[0],[1/L]]
C = [[1, 0],[0,R]]
D = 0
sys = control.ss(A, B, C, D)
print("STATE SPACE MATRICES : \n",sys)

tf = control.ss2tf(sys)
print("Transfer Function : \n", tf)
```

```python
# control.step_response(tf[])
t, y = control.step_response(tf[0,0])
plt.plot(t, y)
# plt.title("Step Response 1")
# plt.xlabel("t")
# plt.ylabel("y")


t, y = control.step_response(tf[1,0])
plt.plot(t, y)
plt.title("Step Response 2")
plt.xlabel("t")
plt.ylabel("y")

plt.legend(["Step Response across Capacitor", "Step Response across Resistor"])
plt.show()
```
